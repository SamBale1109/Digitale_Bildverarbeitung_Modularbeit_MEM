```python
import numpy as np, matplotlib.pyplot as plt
import cv2
```

# Define x(t)


```python
xt = np.zeros(100)
xt[:10] = 1

```

# In Frequenzraum transformieren


```python
Xf = np.fft.fft(xt)
X_magnitude = np.fft.fftshift(np.abs(Xf))
X_phase = np.fft.fftshift(np.angle(Xf))
```

# Plot x(t) and X(f)


```python
# plot x(t) and X(f)
fig, (ax1,ax2,ax3) = plt.subplots(1,3)
ax1.plot(xt)
ax1.set_title("x(t) im Zeitraum")
ax1.set_xlabel("t in samples")
ax1.set_ylabel("Magnitude")

ax2.plot(X_magnitude)
ax2.set_title("X(f) - Amplitude im Frequenzraum")
ax2.set_xlabel("Frequenz in Hz")
ax2.set_ylabel("Amplitude")
ax3.plot(X_phase)
ax3.set_title("X(f) - Phase im Frequenzraum")
ax3.set_xlabel("Frequenz in Hz")
ax3.set_ylabel("Phasenwinkel in radiant")
plt.show()

```

# Load Image and convert to grayscale

```python
img = cv2.imread("testbild.JPG")
x_t = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

X_f = np.fft.fft2(x_t)
size = np.size(X_f)
dims = np.ndim(X_f)
shape = np.shape(X_f)
dtype = X_f.dtype
print(f"X_f: size={size}, dims={dims}, shape={shape}, dtype={dtype}")
X_mag = np.abs(X_f)
X_phase = np.angle(X_f,deg=True)

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

x_axis = np.arange(0,x_t.shape[1],1)
y_axis = np.arange(0,x_t.shape[0],1)
xx,yy = np.meshgrid(x_axis,y_axis)
ax.plot_surface(xx, yy, x_t, rstride=1, cstride=1, cmap='viridis')
```
